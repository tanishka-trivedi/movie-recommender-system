```python
#helper func

def convert(obj):
  l=[]
  for i in ast.literal_eval(obj):
    l.append(i['name'])
  return l
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def convert3(obj):
  l=[]
  counter=0
  for i in ast.literal_eval(obj):
    if counter!=3:
      l.append(i['name'])
      counter+=1
    else:
      break
  return l
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
def convert_director(obj):
  l=[]
  for i in ast.literal_eval(obj):
    if(i['job']=='Director'):
      l.append(i['name'])
      break
  return l
```

```python
movies['crew']=movies['crew'].apply(convert_director)
```

```python
movies.head()

#to show the successful cleaning and pre-processing of data
```

```python
movies['overview'][0]
```

```python
#to make string in list form
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
#make "tag" column

movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new_df=movies[['movie_id','title','tags']]
```

```python
new_df
```

```python
#to convert to string back

new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
#everything to lower case
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

### Vectorization:-

```python
# applying stemming- to avoid similiar words like- accept,accepts, accepted,etc etc like this only
# so they are now counted as one word like all can be brought in the category accept
```

```python
import nltk
```

```python
!pip install nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
s=cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
s
```

```python
list(enumerate(s[0]))

# to keep the indices intact and create list of tupples
```

```python
sorted(list(enumerate(s[0])),reverse=True,key=lambda x:x[1])
# sorted on the basis of 2nd col
```

```python
def recommend(movie):
  mov_in=new_df[new_df['title']==movie].index[0]
  dist=s[mov_in]
  mov_list=sorted(list(enumerate(dist)),reverse=True,key=lambda x:x[1])[1:6]

  for i in mov_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
